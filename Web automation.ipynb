```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import time

options = webdriver.ChromeOptions()

profile = { "plugins.plugins_list": [{"enabled": True, "name": "Chrome PDF Viewer"}],
               "download.default_directory":"C:\\Users\\Dell\\Desktop\\New Folder1", "download.prompt_for_download": False, #To auto download the file
"download.directory_upgrade": True,
"plugins.always_open_pdf_externally": True}
options.add_experimental_option('prefs', profile) #setting up of chrome options 

wd = webdriver.Chrome("C:/Users/Dell/Desktop/chromedriver",options=options)

time.sleep(2)

URL ='https://delhihighcourt.nic.in/'
wd.get(URL)

time.sleep(2)

judgement = wd.find_element_by_link_text('Judgements')
judgement.click()
judgement.click()

time.sleep(2)

PDF_Judgement = wd.find_element_by_link_text('PDF Judgement')
PDF_Judgement.click()

wd.switch_to.window(wd.window_handles[1])

time.sleep(2)

partyname = wd.find_element_by_name('Submit4')
partyname.send_keys(Keys.ENTER)

wd.switch_to.frame("dynfr")

wait = WebDriverWait(wd, 10)
wait.until(EC.presence_of_element_located((By.XPATH, '//input[@name=\"p_name\"]')))

time.sleep(2)

p_name = wd.find_element_by_xpath('//input[@name="p_name"]')
p_name.send_keys(' ')

wd.execute_script("arguments[0].removeAttribute('readonly')", WebDriverWait(wd, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@name='frdate']")))) #changing the default value to readable text
wd.execute_script("arguments[0].removeAttribute('readonly')", WebDriverWait(wd, 10).until(EC.element_to_be_clickable((By.XPATH, "//input[@name='todate']")))) #changing the default value to readable text

wd.find_element_by_xpath("//input[@name='frdate']").clear() #clear values in text box

time.sleep(1)

from_date = wd.find_element_by_id('frdate')
from_date.send_keys('15/09/2021')

time.sleep(2)
wd.find_element_by_xpath("//input[@name='todate']").clear() #clear values in text box

time.sleep(1)
to_date = wd.find_element_by_id('todate')
to_date.send_keys('15/09/2021')

time.sleep(2)
wd.find_element_by_name('Submit').send_keys(Keys.ENTER)
time.sleep(2)

wait.until(EC.presence_of_element_located((By.XPATH, '//a[@target="_blank"]')))
wd.find_element_by_xpath('//a[@target="_blank"]').click()
wd.find_element_by_xpath('//a[@target="_blank"]').click()
```
